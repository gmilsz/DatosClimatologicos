```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('resumenTemperatura/360019.csv')
```

```python
data2004 = data.loc[data.Annio == 2004]
data2008 = data.loc[data.Annio == 2008]
data2012 = data.loc[data.Annio == 2012]
data2016 = data.loc[data.Annio == 2016]
data2020 = data.loc[data.Annio == 2020]
```

```python
plt.plot(range(366), data2004.Tminima)
plt.plot(range(366), data2008.Tminima)
plt.plot(range(366), data2012.Tminima)
plt.plot(range(366), data2016.Tminima)
plt.plot(range(366), data2020.Tminima)

plt.xlabel('Das')
plt.ylabel('')
```
